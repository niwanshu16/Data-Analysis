```python
import seaborn as sns;sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
mu = 0 # Mean
sigma = 1 # Standard deviation

Observations = [random.normalvariate(mu,sigma) for _ in range(100000)]
```

```python
sns.distplot(Observations)

plt.axvline(x = np.mean(Observations)+np.std(Observations) , color = 'y')
plt.axvline(x = np.mean(Observations)-np.std(Observations) , color = 'y') # 65 % of data lies within 1 std

plt.axvline(x = np.mean(Observations)+2*np.std(Observations) , color = 'g')
plt.axvline(x = np.mean(Observations)- 2*np.std(Observations) , color = 'g') #95% data lies within 2 std

plt.show()
```

```python
# Empirical summary using describe method
pd.Series(Observations).describe()
```

```python
sns.boxplot(Observations)
plt.show()
```

```python
# Generate 3 samples from Observations

sampleA = random.sample(Observations,100)
sampleB = random.sample(Observations,100)
sampleC = random.sample(Observations,100)
```

```python
fig , ax = plt.subplots() # Used to plot histogram on top of each other

sns.distplot(sampleA , ax = ax) # Gives the ax =  ax , to identify each histogram where to plot
sns.distplot(sampleB , ax = ax)
sns.distplot(sampleC , ax = ax)

plt.show()
```

### Generate the data for  problem statement of students in college getting sleep on average 7 and standrard deviation 1.7

```python
mu = 7
std = 1.7

Observations = [random.normalvariate(mu,sigma) for _ in range(100000)]

sns.distplot(Observations)
plt.axvline(x = np.mean(Observations) - np.std(Observations) ,color = 'y')
plt.axvline(x = np.mean(Observations) + np.std(Observations) ,color = 'y')

#plt.axvline(np.mean(Observations) + np.std(Observations), 0, .59, color = 'g')
#plt.axvline(np.mean(Observations) - np.std(Observations), 0, .59, color = 'g')

plt.axvline(x = np.mean(Observations) - 2*np.std(Observations) , color = 'g')
plt.axvline(x = np.mean(Observations) + 2*np.std(Observations) , color = 'g')

plt.show()
```

```python
from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(Observations)


plt.plot(ecdf.x , ecdf.y)

plt.axhline(y = 0.025, color = 'g')
plt.axvline(x = np.mean(Observations) - 2*np.std(Observations),color = 'g') # The area below intersection describes that 2.5% of data lies in that region
plt.axvline(x = np.mean(Observations) + 2*np.std(Observations),color = 'g') 

plt.axhline(y = 0.975,color = 'y')
plt.axvline(x = np.mean(Observations) + 2*np.std(Observations) , color = 'y')
plt.axvline(x = np.mean(Observations) - 2*np.std(Observations) , color = 'y')


plt.axhline(y = 0.67,color = 'brown')
plt.axvline(x = np.mean(Observations) + np.std(Observations) , color = 'brown')
plt.axvline(x = np.mean(Observations) - np.std(Observations) , color = 'brown')


plt.show()

# The below graph shows us that in the rectange box 95% data lies , 2.5% data lies above the yellow line 
```
