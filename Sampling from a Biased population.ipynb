```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;sns.set() #for plotting of the histograms
```

```python
# Recreate simulations from video

mean_uofm = 155.0
sd_uofm = 5
mean_gym = 185.0
sd_gym = 5
gymperc_ = .3
totalPopSize = 40000

#Create the two subgroups
uofm_students = np.random.normal(mean_uofm,sd_uofm,int(totalPopSize*(1-gymperc_)))
gym_students = np.random.normal(mean_gym,sd_gym,int(totalPopSize*gymperc_))

# Create the population from the subgroups
population = np.append(uofm_students,gym_students)

# Set up the figure for plotting
plt.figure(figsize = (10,15))

#Plot the UofM students only
plt.subplot(311)
sns.distplot(uofm_students)
plt.title("UofM students only")
plt.xlim([140,200])
plt.axvline(x = np.mean(uofm_students))

#Plot the gym goers
plt.subplot(3,1,2)
sns.distplot(gym_students)
plt.title('Gym goers student only')
plt.axvline(x = np.mean(gym_students))
plt.xlim([140,200])


#plot both university and gym goers

plt.subplot(3,1,3)
sns.distplot(population)
plt.title('Entire population of UofM students')
plt.axvline(x = np.mean(population))
plt.xlim([140,200])
plt.show()
```

# What happens if we sample from entire population

Using probability sampling we can see that the mean of sample is very close to the actual mean 


```python
numberSamp = 5000
sampSize = 50

mean_distribution = np.empty(numberSamp)

for i in range(numberSamp):
    random = np.random.choice(population,sampSize)
    mean_distribution[i] = np.mean(random)
    

plt.figure(figsize=(10,12))
plt.subplot(3,1,1)
sns.distplot(population).set(title = "Full population of UofM students")
plt.axvline(x = np.mean(population))
plt.xlim([140,210])

plt.subplot(3,1,2)
sns.distplot(mean_distribution).set(title = "Mean distribution of the Mean weight of all the UofM students")
plt.axvline(x = np.mean(population))
plt.axvline(x = np.mean(mean_distribution),color = 'black')
plt.xlim([140,210])

plt.show()
```

# What happens if we sample from the biased population

We can see that the mean is very biased toward gym goers

```python
numberSamp = 5000
sampSize = 3

mean_distribution = np.empty(numberSamp)

for i in range(numberSamp):
    random = np.random.choice(gym_students,sampSize)
    mean_distribution[i] = np.mean(random)
    

plt.figure(figsize=(10,12))
plt.subplot(3,1,1)
sns.distplot(population).set(title = "Full population of UofM students")
plt.axvline(x = np.mean(population))
plt.xlim([140,210])

plt.subplot(3,1,2)
sns.distplot(mean_distribution).set(title = "Mean distribution of the Mean weight of all the UofM students")
plt.axvline(x = np.mean(population))
plt.axvline(x = np.mean(mean_distribution),color = 'black')
plt.xlim([140,210])

plt.show()
```
