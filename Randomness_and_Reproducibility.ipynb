## Random Sampling from distribution

From the lecture we know that simple random sampling has the following properties

* Start with known list of N population units , n random samples is selected .
* Each of the sample is equallly likely to be in sample (n/N)
* Estimates of means , proportions based on SRS are **Unbiased**(meaning they are equal to the population average on average)

```python
mu = 0 #Mean
sigma = 1 #Standard deviation
population = [random.normalvariate(mu,sigma) for _ in range(10000)] # Population using normalvariate
```

```python
sampleA = random.sample(population,500) # Sample of sixe 500
sampleB = random.sample(population,500)
```

```python
np.mean(sampleA)
```

```python
np.std(sampleA)
```

```python
np.mean(sampleB)
```

```python
np.std(sampleB)
```

```python
means = [ np.mean(random.sample(population,1000)) for _ in range(100)] # 100 sample of sample size 1000 to calculate average mean

np.mean(means)
```

```python
standard_dev = [np.std(random.sample(population,1000)) for _ in range(100)] # 100 sample of sample size 1000 to calculate average std

np.mean(standard_dev)
```

```python
sns.distplot(means)
```

```python
sns.distplot(standard_dev)
```
